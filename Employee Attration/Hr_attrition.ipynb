## Train Test Split

```python
#importing library
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
hr.head()
```

### Feature scaling

```python
#importing standardscale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x_train,y_train)
sc.transform(x_test)

```

```python
x_train
```

```python
x_test
```

## Model Building

### 1. Logistic Refrestion

```python
#importing libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
y_pred_lg=lg.predict(x_test)
print("classification Report: ",classification_report(y_test,y_pred_lg))
print("accuracy_score: ",accuracy_score(y_test,y_pred_lg))
print("Confusion matrix: ")
confusion_matrix(y_test,y_pred_lg)
```

Since we are getting 100 % accuracy but there is imbalance 

Balancing imbalance target column. This is achieved using SMOTE technqiue

```python
#Balancing
from imblearn.over_sampling import SMOTE
#x1=sample y1=target
x1,y1 = SMOTE().fit_resample(x,y)
```

```python
x1.shape
```

```python
pd.value_counts(y1).plot.bar()
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1,test_size=0.2,random_state=101)
```

```python
lg=LogisticRegression()
lg.fit(x1_train,y1_train)
y1_pred_lg=lg.predict(x1_test)
print("classification Report: ",classification_report(y1_test,y1_pred_lg))
print("accuracy_score: ",accuracy_score(y1_test,y1_pred_lg))
print("Confusion matrix: ")
confusion_matrix(y1_test,y1_pred_lg)
```

### 2. Decision tree

```python
#checking with decision tree
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x1_train,y1_train)
y1_pred_dt=dt.predict(x1_test)
print("classification Report: ",classification_report(y1_test,y1_pred_dt))
print("accuracy_score: ",accuracy_score(y1_test,y1_pred_dt))
print("Confusion matrix: ")
confusion_matrix(y1_test,y1_pred_dt)
```

### 3. Random Forest

```python
#checking with decision tree
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='entropy', n_estimators=10, random_state=0)
rfc.fit(x1_train,y1_train)
y1_pred_rfc=rfc.predict(x1_test)
print("classification Report: ",classification_report(y1_test,y1_pred_rfc))
print("accuracy_score: ",accuracy_score(y1_test,y1_pred_rfc))
print("Confusion matrix: ")
confusion_matrix(y1_test,y1_pred_rfc)
```
